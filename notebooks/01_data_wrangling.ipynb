```python
import pandas as pd
import kagglehub
from pathlib import Path
import shutil
```

```python
# Download the dataset and get the path to the files, load the data into a DataFrame
path = kagglehub.dataset_download("eoinamoore/historical-nba-data-and-player-box-scores")

print("Path to dataset files:", path)
src = Path(path)
dst = ".." / Path("data") / "raw"

if dst.exists():
    shutil.rmtree(dst)
dst.parent.mkdir(parents=True, exist_ok=True)
shutil.copytree(src, dst)
print(f"Copied {src} to {dst}")

df = pd.read_csv(dst / "PlayerStatistics.csv")
```

```python
# data cleanup
df = df.drop(columns=["gameType", "gameLabel", "gameSubLabel", "seriesGameNumber"])
df = df.dropna(subset=["numMinutes", "points", "reboundsTotal", "assists"])
df = df[df["numMinutes"] > 0]
```

```python
# getting types right
# df["gameDateTimeEst"] = pd.to_datetime(df["gameDateTimeEst"], format="ISO8601")
# df["personId"] = df["personId"].astype(int)
# getting types right
df["gameDateTimeEst"] = pd.to_datetime(df["gameDateTimeEst"], utc=True)
df["personId"] = df["personId"].astype(int)
df = df.sort_values("gameDateTimeEst")
```
